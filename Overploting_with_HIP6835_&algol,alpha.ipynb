```
algol = pd.read_csv('msc-2018-20/aavso_data/aavsodata_algol.txt',usecols=['JD','Magnitude'])
algol.head()
```

```

algol = algol.convert_objects(convert_numeric=True)
algol.dtypes
```

```
summer = algol[['JD','Magnitude']][2000:5000]
summer.plot(y='Magnitude',x='JD',kind='scatter')
plt.gca().invert_yaxis()
```

```
summer['diffMagnitude'] = summer['Magnitude'] - summer['Magnitude'].shift(1)
summer['diffJD'] = summer['JD'] - summer['JD'].shift(1)
summer.head()
```

```
HIP6835 = pd.read_csv('msc-2018-20/aavso_data/HIP6835.txt',usecols=['JD','Magnitude'])
HIP6835.head()
```

```

HIP6835= HIP6835.convert_objects(convert_numeric=True)
HIP6835.dtypes
```

```
winter = HIP6835[['JD','Magnitude']][:150]
autumn = HIP6835[['JD','Magnitude']][500:1000]
rainy = HIP6835[['JD','Magnitude']][2000:5000]
spring = HIP6835[['JD','Magnitude']][6500:9500]
```

```
winter.plot(y='Magnitude',x='JD',kind='scatter')
plt.gca().invert_yaxis()
autumn.plot(y='Magnitude',x='JD',kind='scatter')
plt.gca().invert_yaxis()
rainy.plot(y='Magnitude',x='JD',kind='scatter')
plt.gca().invert_yaxis()
spring.plot(y='Magnitude',x='JD',kind='scatter')
plt.gca().invert_yaxis()
```

```
rainy = HIP6835[['JD','Magnitude']][2000:5000]
rainy.plot(y='Magnitude',x='JD',kind='scatter')
plt.gca().invert_yaxis()
```

```
rainy['diffMagnitude'] = rainy['Magnitude'] - rainy['Magnitude'].shift(1)
rainy['diffJD'] = rainy['JD'] - rainy['JD'].shift(1)
rainy.head()
```

```
import plotly.graph_objects as go
import numpy as np




fig = go.Figure()

fig.add_trace(go.Scatter(
    x=summer['diffJD'], y=summer['diffMagnitude'],
    name='algol',
    mode='markers',
    marker_color='rgba(152, 0, 0, .8)'
))

fig.add_trace(go.Scatter(
    x=rainy['diffJD'], y=rainy['diffMagnitude'],
    name='HIP6835',
    marker_color='rgba(255, 182, 193, .9)'
))


# Set options common to all traces with fig.update_traces
fig.update_traces(mode='markers', marker_line_width=2, marker_size=10)
fig.update_layout(xaxis_type="log",yaxis_type="log",title='Styled Scatter',
                  yaxis_zeroline=False, xaxis_zeroline=False)


fig.show()
```
